# Subplot in Matplotlib

**ðŸ“˜ What is a Subplot?**
A subplot is a way to show multiple plots in a single figure.
Instead of creating one chart at a time, you can organize many in rows and columns using subplots.

**âœ… Why Use Subplots?**
âœ” Compare multiple charts side by side
âœ” Save space in your visualizations
âœ” Useful in dashboards, reports, and data analysis

# ðŸ“Œ Functions for Subplots in Matplotlib

* plt.subplot(nrows, ncols, index)  
* plt.subplots(nrows=1, ncols=1, **kwargs)  
* fig.add_subplot()  
* fig, ax = plt.subplots()  
* ax.plot(), ax.bar(), ax.scatter(), etc.  
* plt.tight_layout()  
* plt.show()  

# ðŸ“Œ Parameters for plt.subplot() and plt.subplots()

* nrows             â€ƒâ€ƒâ€ƒâ€ƒ# number of rows  
* ncols             â€ƒâ€ƒâ€ƒâ€ƒ# number of columns  
* index             â€ƒâ€ƒâ€ƒâ€ƒ# position of the current subplot (starts at 1)  

# ðŸ“Œ Parameters for plt.subplots() (with fig, ax)

* figsize           â€ƒâ€ƒâ€ƒâ€ƒ# width and height of the figure  
* sharex / sharey   â€ƒâ€ƒâ€ƒâ€ƒ# share x/y axis between subplots (True/False)  
* constrained_layout â€ƒâ€ƒâ€ƒâ€ƒ# auto-adjust layout (like tight_layout)  
* dpi               â€ƒâ€ƒâ€ƒâ€ƒ# dots per inch (resolution)  

# ðŸ“Œ Common Methods Used on Subplot Axes (ax)

* ax.plot(x, y)           â€ƒâ€ƒâ€ƒâ€ƒ# line plot  
* ax.bar(x, height)       â€ƒâ€ƒâ€ƒâ€ƒ# bar plot  
* ax.scatter(x, y)        â€ƒâ€ƒâ€ƒâ€ƒ# scatter plot  
* ax.hist(x)              â€ƒâ€ƒâ€ƒâ€ƒ# histogram  
* ax.set_title()          â€ƒâ€ƒâ€ƒâ€ƒ# set title for individual subplot  
* ax.set_xlabel()         â€ƒâ€ƒâ€ƒâ€ƒ# set x-axis label  
* ax.set_ylabel()         â€ƒâ€ƒâ€ƒâ€ƒ# set y-axis label  
* ax.grid(True/False)     â€ƒâ€ƒâ€ƒâ€ƒ# show/hide grid  
* ax.legend()             â€ƒâ€ƒâ€ƒâ€ƒ# show legend  


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
```

```python
plt.plot(study_hours,marks,'bo')
```

```python
plt.plot(study_hours,marks,'bo')
```

```python
plt.plot(study_hours,marks,'y-')
plt.plot(study_hours,marks,'bo')
```

```python
plt.bar(study_hours,marks,color = 'red',linestyle = '-',edgecolor = 'green',linewidth = 2)
```

```python
plt.figure(figsize = (8,8))

plt.subplot(2,2,1)
plt.bar(study_hours,marks,color = 'red',linestyle = '-',edgecolor = 'green',linewidth = 2)

plt.subplot(2,2,2)
plt.plot(study_hours,marks,'bo')

plt.subplot(2,2,3)
plt.plot(study_hours,marks,'y-')
plt.plot(study_hours,marks,'bo')

plt.subplot(2,2,4)
plt.plot(study_hours,marks,'bo')
```
