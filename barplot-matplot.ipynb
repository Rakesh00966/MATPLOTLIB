# Barplot
* A bar plot is used to compare different categories using rectangular bars.
* Each barâ€™s height (or length) represents the value of that category.

# ðŸ“Œ Functions for Bar Plot in Matplotlib

* plt.bar(x, height, **kwargs)  
* plt.barh(y, width, **kwargs)  
* plt.xlabel()  
* plt.ylabel()  
* plt.title()  
* plt.grid()  
* plt.legend()  
* plt.xticks()  
* plt.yticks()  
* plt.xlim()  
* plt.ylim()  
* plt.text()  
* plt.tight_layout()  
* plt.show()  


# ðŸ“Œ Parameters for plt.bar() and plt.barh()

* x / y             â€ƒâ€ƒâ€ƒâ€ƒ# categories or positions on x/y axis  
* height / width    â€ƒâ€ƒâ€ƒâ€ƒ# height (for vertical bars) or width (for horizontal bars)  
* color             â€ƒâ€ƒâ€ƒâ€ƒ# fill color of bars  
* edgecolor         â€ƒâ€ƒâ€ƒâ€ƒ# border color of bars  
* linewidth         â€ƒâ€ƒâ€ƒâ€ƒ# thickness of border  
* alpha             â€ƒâ€ƒâ€ƒâ€ƒ# transparency (0 to 1)  
* label             â€ƒâ€ƒâ€ƒâ€ƒ# label for legend  
* width             â€ƒâ€ƒâ€ƒâ€ƒ# width of each bar (in plt.bar)  
* align             â€ƒâ€ƒâ€ƒâ€ƒ# 'center' (default) or 'edge'  
* bottom / left     â€ƒâ€ƒâ€ƒâ€ƒ# starting point of the bar (for stacked bars)  
* tick_label        â€ƒâ€ƒâ€ƒâ€ƒ# custom labels for ticks  
* hatch             â€ƒâ€ƒâ€ƒâ€ƒ# pattern on bars ('/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*')  
* zorder            â€ƒâ€ƒâ€ƒâ€ƒ# layering order (higher is on top)  



# ðŸ“Œ Parameters for other functions

* rotation          â€ƒâ€ƒâ€ƒâ€ƒ# used in plt.xticks() / plt.yticks() to rotate labels  
* fontsize          â€ƒâ€ƒâ€ƒâ€ƒ# used in labels, titles, ticks, legend  
* loc               â€ƒâ€ƒâ€ƒâ€ƒ# used in plt.legend(), plt.title() ('upper right', 'center', etc.)  
* linestyle         â€ƒâ€ƒâ€ƒâ€ƒ# used in plt.grid() ('--', '-', ':', etc.)  
* alpha             â€ƒâ€ƒâ€ƒâ€ƒ# used in grid and bars for transparency  
* ha / va           â€ƒâ€ƒâ€ƒâ€ƒ# horizontalalignment / verticalalignment (used in plt.text)  


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks = [89, 90, 45, 78, 99]
plt.bar(subjects,marks)
```

```python
plt.bar(subjects,marks,color = 'red')
```

```python
colours = ['orange','grey','green']
plt.bar(subjects,marks,color = colours)
```

```python
plt.bar(subjects,marks,color = colours, linestyle = '-',linewidth = 3, edgecolor = 'blue')
```

```python
plt.bar(subjects,marks,color = colours, linestyle = '-.',linewidth = 3, edgecolor = 'blue')
```

```python
plt.bar(subjects,marks,color = colours, linestyle = '-',linewidth = 3, edgecolor = 'blue',width = 0.5)
```

```python
plt.barh(subjects,marks,color = colours, linestyle = '-',linewidth = 3, edgecolor = 'blue')
```

```python
subjects_len = np.arange(len(subjects))
subjects_len
```

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks1 = [89, 90, 45, 78, 99]
marks2 = [98,65,78,66,20]
plt.bar(subjects,marks1,color = colours, alpha = 0.5)
plt.bar(subjects,marks2,color = colours,alpha = 0.5)
```

```python
subjects_len = np.arange(len(subjects))
width = 0.4
plt.bar(subjects,marks1,color = colours, width = width)
plt.bar(subjects_len+width,marks2,color = colours,width = width ,alpha = 0.5)
```
