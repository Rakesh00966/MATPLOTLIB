# Histogram
**ðŸ“Š What is a Histogram?**
A histogram is a graph that shows the distribution of a numerical variable.
It divides values into intervals (called bins) and shows how many data points fall into each bin.

ðŸ‘‰ Example: Showing how many students scored between 40â€“50, 50â€“60, etc.

# ðŸ“Š Bar Plot vs ðŸ“‰ Histogram â€“ Key Differences

"""
ðŸ“Œ Type of Data:
ðŸ”¹ Bar Plot       â†’ Categorical data (e.g., Product, Gender)
ðŸ”¹ Histogram       â†’ Numerical continuous data (e.g., Age, Marks)

ðŸ“Œ Purpose:
ðŸ”¹ Bar Plot       â†’ Compare values between different categories
ðŸ”¹ Histogram       â†’ Show the distribution of data over intervals (bins)

ðŸ“Œ Bars Represent:
ðŸ”¹ Bar Plot       â†’ Each bar = one category
ðŸ”¹ Histogram       â†’ Each bar = a range of values

ðŸ“Œ Gaps Between Bars:
ðŸ”¹ Bar Plot       â†’ Bars are separated (with gaps)
ðŸ”¹ Histogram       â†’ Bars touch each other (no gaps)

ðŸ“Œ X-Axis Meaning:
ðŸ”¹ Bar Plot       â†’ Categories (like "Shoes", "Shirts")
ðŸ”¹ Histogram       â†’ Value ranges (like 60â€“70, 70â€“80)

ðŸ“Œ Common Use Cases:
ðŸ”¹ Bar Plot       â†’ Sales by product, Population by city
ðŸ”¹ Histogram       â†’ Marks distribution, Age distribution
"""

# ðŸ“Œ Parameters for plt.hist()

* x                â€ƒâ€ƒâ€ƒâ€ƒ# input data  
* bins             â€ƒâ€ƒâ€ƒâ€ƒ# number of bins or custom bin edges (int or list)  
* color / c        â€ƒâ€ƒâ€ƒâ€ƒ# fill color of bars  
* edgecolor        â€ƒâ€ƒâ€ƒâ€ƒ# border color of bars  
* linewidth        â€ƒâ€ƒâ€ƒâ€ƒ# thickness of border  
* alpha            â€ƒâ€ƒâ€ƒâ€ƒ# transparency (0 to 1)  
* label            â€ƒâ€ƒâ€ƒâ€ƒ# label for legend  
* orientation      â€ƒâ€ƒâ€ƒâ€ƒ# 'vertical' (default) or 'horizontal'  
* histtype         â€ƒâ€ƒâ€ƒâ€ƒ# 'bar', 'barstacked', 'step', 'stepfilled'  
* cumulative       â€ƒâ€ƒâ€ƒâ€ƒ# if True, shows cumulative histogram  
* density          â€ƒâ€ƒâ€ƒâ€ƒ# if True, shows probability density instead of counts  
* stacked          â€ƒâ€ƒâ€ƒâ€ƒ# if True, stacks multiple datasets  
* rwidth           â€ƒâ€ƒâ€ƒâ€ƒ# relative width of bars (0 to 1)  
* zorder           â€ƒâ€ƒâ€ƒâ€ƒ# layering order  


# ðŸ“Œ Parameters for other functions

* rotation         â€ƒâ€ƒâ€ƒâ€ƒ# used in plt.xticks() / plt.yticks()

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
students_marks = np.random.randint(0,100,(50))
students_marks
```

```python
plt.hist(students_marks)
```

```python
plt.hist(students_marks)
plt.xticks(np.arange(0,100,5))
```

```python
bins = np.arange(0,100,5)
plt.hist(students_marks,bins =bins, color = 'orange')
plt.xticks(np.arange(0,100,5))
```

```python
plt.hist(students_marks,color = 'red', orientation = 'horizontal')
plt.yticks(np.arange(0,100,5))
```

```python
plt.hist(students_marks,color = 'blue', linestyle = '-',edgecolor = 'yellow', linewidth = 5)
plt.xticks(np.arange(0,100,5))
```
