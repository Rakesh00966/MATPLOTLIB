# üìò Simple Explanation of Pie Chart
A pie chart is a circular chart divided into slices to show proportions of a whole.
Each slice represents a category‚Äôs percentage or share in the total.

# üìå When to Use a Pie Chart?
‚úÖ When you want to show percentage contribution of different categories
‚úÖ When you have limited categories (ideally less than 6)
‚ùå Not good for comparing exact values or when you have too many categories

# üìå Function for Pie Chart in Matplotlib

* plt.pie(x, **kwargs)  
* plt.title()  
* plt.legend()  
* plt.tight_layout()  
* plt.show()  

# üìå Parameters for plt.pie()

* x                ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# list or array of values for each slice  
* labels           ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# names of each category  
* colors           ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# custom colors for slices  
* autopct          ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# format for showing percentage on slices (e.g., '%1.1f%%')  
* startangle       ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# angle to start the pie chart (default is 0)  
* shadow           ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# adds shadow behind slices (True/False)  
* explode          ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# pushes a slice out (e.g., [0.1, 0, 0, 0])  
* radius           ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# controls the size of the pie  
* counterclock     ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# draws slices in counterclockwise direction (default is True)  
* wedgeprops       ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# dictionary to customize wedges (e.g., edgecolor, linewidth)  
* labeldistance    ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# distance of label from center  
* pctdistance      ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# distance of percentage text from center  
* frame            ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# if True, adds a frame around pie  


# üìå Parameters for Other Functions

* fontsize         ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# used in title, legend, etc.  
* loc              ‚ÄÉ‚ÄÉ‚ÄÉ‚ÄÉ# position of legend (used in plt.legend())  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12,10))
subjects = ['Phy','Soc','Mat','Sci']
marks = [10,90,84,88]
plt.pie(marks,labels = subjects)
```

```python
subjects = ['Phy','Soc','Mat','Sci']
marks = [10,90,84,88]
plt.pie(marks,labels = subjects,radius = 0.2)
```

```python
plt.pie(marks,labels = subjects,radius = 2)
```

```python
colours = ['red','orange','green','blue']
plt.pie(marks,labels = subjects,radius = 2, colors = colours)
```

```python
colours = ['red','orange','green','blue']
plt.pie(marks,labels = subjects,radius = 2, colors = colours,autopct = '%0.1f%%')
plt.show()
```

```python
explode_values = [0,0.1,0,0]
plt.pie(marks,labels = subjects,radius = 2, colors = colours,autopct = '%0.1f%%',explode =explode_values)
```

# Text props and wedge props

```python
text_props = {'color':'k',
             'fontsize': 22}
wedge_props = {'linestyle':'-.','edgecolor':'k'}
plt.pie(marks,labels = subjects,radius = 2, colors = colours,autopct = '%0.1f%%',wedgeprops = wedge_props,textprops = text_props)
```

```python
text_props = {'color':'k',
             'fontsize': 22}
wedge_props = {'linestyle':'-.','edgecolor':'k'}
plt.pie(marks,labels = subjects,radius = 2, colors = colours,autopct = '%0.1f%%',wedgeprops = wedge_props,textprops = text_props)
plt.legend()
```
