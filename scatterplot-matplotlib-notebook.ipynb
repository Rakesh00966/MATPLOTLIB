# Scatterplot
* Scatter plot is a type of graph which shows corelation between two variables using dots.
* Each dot represents corelation between two variables having x,y values

**Basics of Scatterplot**
* Used to show correlation or distribution.
* X-axis: independent variable
* Y-axis: dependent variable

# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Defing Data

```python
x = [1,2,3,4,5]
y = [10,20,33,44,55]
```

# Implementing Basic Scatterplot

```python
plt.scatter(x,y)
plt.show()
```

# Basic Designing

```python
#Changing markers colour and shape
plt.scatter(x,y,color= 'red',marker = 'v')
plt.show()
```

# Defing plt.figure()
* It is used open a blank sheet like opening a new sheet for drawing

```python
a  = plt.figure()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y,color = 'red',marker= 'v')
```

# **Using Plot()**

```python
plt.figure(figsize= (8,8))
plt.plot(x,y,'bv',markersize = 10)
```

# Multiple plots on same scatterplot

```python
temperature_delhi = [39, 42, 35, 33, 30, 28, 37, 40, 29, 31]
humidity_delhi = [22, 20, 25, 27, 30, 35, 24, 23, 32, 28]

temperature_hyderabad = [32, 34, 33, 30, 28, 27, 31, 35, 29, 26]
humidity_hyderabad = [45, 42, 44, 46, 48, 50, 43, 41, 49, 47]
```

```python
plt.figure(figsize=(12,10))
plt.plot(temperature_delhi,humidity_delhi,'ro')
```

# Use of range()
* xticks,yticks are used to access x axis and y axis

```python
plt.figure(figsize=(12,10))
plt.xticks(np.arange(0,60,5))
plt.yticks(np.arange(0,60,5))
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title("Temp vs Humd")
plt.plot(temperature_delhi,humidity_delhi,'ro')
plt.plot(temperature_hyderabad,humidity_hyderabad,'bo')
plt.show()
```
